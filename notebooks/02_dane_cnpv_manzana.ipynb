# 02 — Exploración DANE CNPV 2018 a nivel de manzana

Objetivo: cuantificar qué variables del Censo 2018 están disponibles
a nivel de manzana censal para los municipios de Urabá, y qué porcentaje
tiene supresión de privacidad (manzanas con < 3 unidades estadísticas).

**Fuentes:**
- Geovisor DANE CNPV 2018: https://geoportal.dane.gov.co/geovisores/sociedad/cnpv2018-detallado/
- Microdatos: https://microdatos.dane.gov.co/index.php/catalog/643
- MGN 2024: https://www.arcgis.com/home/item.html?id=da6856c2040c4098831605384715d35b

**Pregunta clave:** ¿Qué % de manzanas tienen supresión? ¿Es necesario usar
interpolación areal (tobler) para bajar de sector a manzana?

```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
import sys; sys.path.append('..')

# Códigos DIVIPOLA de Urabá
DIVIPOLA_URABA = ['05045', '05147', '05120', '05837', '05544', '05665', '05659', '05051']

print('Descarga el MGN 2024 desde:')
print('https://www.arcgis.com/home/item.html?id=da6856c2040c4098831605384715d35b')
print('y guárdalo en: data/raw/dane/mgn_manzanas_2024.shp')
```

```python
# Cargar MGN cuando esté disponible
mgn_path = Path('../data/raw/dane/mgn_manzanas_2024.shp')

if mgn_path.exists():
    mgn = gpd.read_file(mgn_path)
    # Filtrar Urabá
    col_mpio = [c for c in mgn.columns if 'MPIO' in c.upper() and 'CD' in c.upper()][0]
    uraba = mgn[mgn[col_mpio].isin(DIVIPOLA_URABA)]
    print(f'Manzanas Urabá: {len(uraba):,}')
    print(f'Municipios: {uraba[col_mpio].value_counts().to_dict()}')
    uraba.plot(figsize=(12, 10), column=col_mpio, legend=True, cmap='Set3')
    plt.title('Manzanas censales — Urabá')
else:
    print('[pendiente] Descargar MGN para continuar')
```
