# 01 — Auditoría de cobertura OSM en Urabá

Objetivo: cuantificar cuántos datos de red vial y equipamientos existen
en OpenStreetMap para cada municipio de Urabá antes de decidir qué fuentes
alternativas (IGAC, REPS, SIMAT) son necesarias.

**Pregunta clave:** ¿Es la cobertura OSM suficiente para análisis de accesibilidad
peatonal, o necesitamos complementar con catastro multipropósito IGAC?

```python
import osmnx as ox
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
import sys; sys.path.append('..')
from src.ingestion.osm_extractor import MUNICIPIOS_URABA, auditar_cobertura_osm, AMENITY_TAGS

ox.settings.log_console = False
```

```python
# Auditar un municipio piloto primero
municipio_piloto = 'Apartadó, Antioquia, Colombia'
reporte = auditar_cobertura_osm(municipio_piloto)
print(reporte)
```

```python
# Visualizar red vial de Apartadó
G = ox.graph_from_place(municipio_piloto, network_type='walk')
fig, ax = ox.plot_graph(G, figsize=(12, 10), node_size=3, edge_linewidth=0.5,
                         bgcolor='white', node_color='steelblue', edge_color='#333')
ax.set_title(f'Red vial peatonal — {municipio_piloto}', fontsize=14)
plt.tight_layout()
```

```python
# Auditar todos los municipios de Urabá
reportes = []
for m in MUNICIPIOS_URABA:
    try:
        r = auditar_cobertura_osm(m)
        reportes.append(r)
        print(f'✓ {m.split(",")[0]}')
    except Exception as e:
        print(f'✗ {m.split(",")[0]}: {e}')
        reportes.append({'municipio': m, 'error': str(e)})

df_audit = pd.DataFrame(reportes)
df_audit
```
